```python
import matplotlib as mpl ; mpl.rcParams['axes.unicode_minus'] = False
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
data1=pd.read_csv('/content/gdrive/My Drive/quiz1.csv')
data1
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.grid(True, color='gray', linestyle='-', linewidth=1, axis='y')
ax.set_yticks(range(0,80,10))
ax.bar(data1['일'], data1['해외유입'], label='해외유입',color='r')
ax.bar(data1['일'], data1['국내발생'], bottom=data1['해외유입'], label='국내발생',color='b')
ax.set_title('감염경로구분에 따른 신규확진자 현황')
ax.legend(bbox_to_anchor=(0.55, -0.08))
plt.show()
```

```python
data2 = pd.read_csv('/content/gdrive/My Drive/quiz2.csv')
data2
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.grid(True, color='gray', linestyle='-', linewidth=1, axis='y')
ax.set_yticks(range(0,18000,2000))
ax.bar(data2['일'], data2['누적 확진환자'], label='누적 확진환자',color = 'b')
ax.set_ylabel('누적 확진환자(명)')
ax2 = ax.twinx()
ax2.plot(data2['일'], data2['일 확진환자'], linestyle='-', marker='o', color='y',label='일 확진환자')
ax2.set_yticks(range(0,180,20))
ax2.set_ylabel('일 확진환자(명)')
ax.set_title('확진환자 내 일일 및 누적 격리해제 추세')
ax.legend(bbox_to_anchor=(0.55, -0.08))
ax2.legend(bbox_to_anchor=(0.75, -0.08))
plt.show()
```
